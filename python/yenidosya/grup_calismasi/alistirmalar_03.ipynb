```python
def yeterlipuanlarisayma(notlarlistesi):
    """
    :param notlarlistesi: Een lijst van tentamennotlar
    :return: Het aantal notlar in de array dat groter is dan 5.5
    """
    yeterlipuansayisi= 0
    for puan in notlarlistesi:
        if puan >= 5.5:
            yeterlipuansayisi+= 1 # een korte manier om een variabele 1 op te hogen
    return yeterlipuansayisi


def alinanenyuksekpuan(notlarlistesi):
    """
    :param notlarlistesi: Een array van tentamennotlar
    :return: het hoogste not van de array
    """
    # Zet het maximum op 0 aan het begin van de functie
    maxnot = 0
    for puan in notlarlistesi:
        # Als het not in de lijst groter is dan het maximum
        if puan > maxnot:
            # Maak het maximum dan gelijk aan het not dat je controleert
            maxnot = puan
    # Als je de hele array gehad hebt, is maxnot gelijk aan het maximum
    # van de hele array
    return maxnot

def ortalamahesabi(notlarlistesi):
    """
    :param notlarlistesi: Een array van tentamennotlar
    :return: Het gemiddelde van alle notlar in de notlarlistesi
    """
    toplam= 0
    for puan in notlarlistesi:
      toplam= toplam+ puan
      gemiddelde = toplam/ len(notlarlistesi)
    return gemiddelde

# HIER BEGINT HET HOOFDPROGRAMMA

# Vraag het aantal notlar dat de gebruiker wil invoeren
notlarinsayisi = int(input("Hoeveel notlar wilt u invoeren? "))

# Maak een lijst aan en vul deze met de invoer van de gebruiker
notlar = []
for teller in range(notlarinsayisi):
    girilennotlar = float(input(f"not student {teller + 1}: "))
    notlar.append(girilennotlar)

# Print het aantal notlar, het gemiddelde, het aantal voldoendes en
# en het hoogste not
print()
print("Aantal notlar    : ", notlarinsayisi)
print("Gemiddelde not : ", round(ortalamahesabi(notlar),2))  # Dit kan ook: sum(notlar) / len(notlar))
print("Aantal voldoendes : ", yeterlipuanlarisayma(notlar))
print("Hoogste not    : ", alinanenyuksekpuan(notlar))  # Dit kan ook: max(notlar)

```
